```python
import pandas as pd
import os

from datetime import datetime
```

```python
#read the raw data
file = pd.read_csv('file.csv', usecols = ['date', 'user', 'pc', 'activity', 'to_removable_media', 'from_removable_media'])
device = pd.read_csv('device.csv', usecols = ['date', 'user', 'pc', 'activity'])
```

```python
#necessary functions

#return just date
def split_date(datetime):
    return datetime.date()

#return just time
def split_time(datetime):
    return datetime.time()

#determine if actions are performed in working hours
def is_working_hours(timestamp):
    time = timestamp.time()
    
    start = datetime.strptime('09:00:00', '%H:%M:%S').time()
    end = datetime.strptime('17:00:00', '%H:%M:%S').time()
    
    # Check if it's a weekday (Monday: 0, Tuesday: 1, ..., Friday: 4)
    if timestamp.weekday() in range(0, 5):
        if (time >= start and time <= end): # Check if it's 9 AM or later
            return "Work Hours"
        return "Off Hours"
    else:
        return 'Weekend'  # It's not a weekday
```

```python
#modify file data
#turn date into datetime
file['date'] = pd.to_datetime(file.date, format = '%m/%d/%Y %H:%M:%S')

#split dates and times
file['only_date'] = file['date'].apply(split_date)
file['only_time'] = file['date'].apply(split_time)

file['WorkingHours'] = file.date.apply(is_working_hours)
```

```python
#group dataframe by date
grouped = file.groupby('only_date')
```

```python
#hold the file names to make a log file later
file_names = []

#create separate csv's for each date
for date, group in grouped:
    current_file_name = f'file-{date}.csv'
    group.to_csv(f'file_data/{current_file_name}', index = False)
    file_names.append(current_file_name)
```

```python
#create a log file with the file names to make access easier
with open('file_data/file_files.log', 'w') as file:
    for name in file_names:
        file.write(name + '\n')
```

```python
#modify device data
#turn date into datetime
device['date'] = pd.to_datetime(device.date, format = '%m/%d/%Y %H:%M:%S')

#split dates and times
device['only_date'] = device['date'].apply(split_date)
device['only_time'] = device['date'].apply(split_time)

device['WorkingHours'] = device.date.apply(is_working_hours)
```

```python
#group dataframe by date
grouped = device.groupby('only_date')
```

```python
#hold the file names to make a log file later
file_names = []

#create separate csv's for each date
for date, group in grouped:
    current_file_name = f'device-{date}.csv'
    group.to_csv(f'device_data/{current_file_name}', index = False)
    file_names.append(current_file_name)
```

```python
#create a log file with the file names to make access easier
with open('device_data/device_files.log', 'w') as file:
    for name in file_names:
        file.write(name + '\n')
```
